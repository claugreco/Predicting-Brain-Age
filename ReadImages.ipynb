```python
%matplotlib inline
import numpy as np
import nibabel as nib
import glob
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x_data = []
files = sorted(glob.glob('/Users/claugreco/Documents/Capstone/Data/IXI/IXI-T1/*.nii.gz'))
print(len(files))
for f in files:
    img = nib.load(f)
    img_data = img.get_data()
    img_data = np.asarray(img_data)  
    img_data.astype(int)             
    img_data = img_data[130, :, :]
#     plt.imshow(img_data, cmap='gray')
#     plt.show()
    x_data.append(list(img_data.flatten()))
print(len(x_data))
```
